#DFS

```python
def dfs(graph, start_node, target_node):
    stack = [(start_node, [start_node])]
    visited = set()

    print(f"\nMemulai DFS dari {start_node} mencari {target_node}\n")

    while stack:
        current_node, path = stack.pop()
        print(f"Mengunjungi: {current_node}")

        if current_node in visited:
            continue

        visited.add(current_node)

        if current_node == target_node:
            print("\nGoal ditemukan!")
            return path

        for neighbor in reversed(graph.get(current_node, [])):
            if neighbor not in visited:
                stack.append((neighbor, path + [neighbor]))
                print(f"  ➜ Tambah ke stack: {neighbor}")

    print("\nGoal tidak ditemukan.")
    return None


# =========================
# GRAF A SAMPAI Z
# =========================

graph = {
    'A': ['B', 'C', 'D'],
    'B': ['E', 'F'],
    'C': ['G', 'H'],
    'D': ['I'],
    'E': [],
    'F': ['J'],
    'G': [],
    'H': [],
    'I': ['K'],
    'J': [],
    'K': ['L'],
    'L': ['M'],
    'M': ['N'],
    'N': ['O'],
    'O': ['P'],
    'P': ['Q'],
    'Q': ['R'],
    'R': ['S'],
    'S': ['T'],
    'T': ['U'],
    'U': ['V'],
    'V': ['W'],
    'W': ['X'],
    'X': ['Y'],
    'Y': ['Z'],
    'Z': []
}

# =========================
# EKSEKUSI DFS
# =========================

path = dfs(graph, 'A', 'Z')

if path:
    print("\nJalur dari A ke Z:")
    print(" → ".join(path))
```

#BFS

```python
from collections import deque

def bfs(graph, start_node, target_node):
    queue = deque([(start_node, [start_node])])
    visited = set()

    print(f"\nMemulai BFS dari {start_node} mencari {target_node}\n")

    while queue:
        current_node, path = queue.popleft()
        print(f"Mengunjungi: {current_node}")

        if current_node in visited:
            continue

        visited.add(current_node)

        if current_node == target_node:
            print("\nGoal ditemukan!")
            return path

        for neighbor in graph.get(current_node, []):
            if neighbor not in visited:
                queue.append((neighbor, path + [neighbor]))
                print(f"  ➜ Tambah ke antrean: {neighbor}")

        print("-" * 40)

    print("\nGoal tidak ditemukan.")
    return None


# =========================
# GRAF A SAMPAI Z
# =========================

graph = {
    'A': ['B', 'C', 'D'],
    'B': ['E', 'F'],
    'C': ['G', 'H'],
    'D': ['I'],
    'E': [],
    'F': ['J'],
    'G': [],
    'H': [],
    'I': ['K'],
    'J': [],
    'K': ['L'],
    'L': ['M'],
    'M': ['N'],
    'N': ['O'],
    'O': ['P'],
    'P': ['Q'],
    'Q': ['R'],
    'R': ['S'],
    'S': ['T'],
    'T': ['U'],
    'U': ['V'],
    'V': ['W'],
    'W': ['X'],
    'X': ['Y'],
    'Y': ['Z'],
    'Z': []
}

# =========================
# EKSEKUSI BFS
# =========================

path = bfs(graph, 'A', 'Z')

if path:
    print("\nJalur terpendek dari A ke Z:")
    print(" → ".join(path))
```
